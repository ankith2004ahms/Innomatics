```python
# Problem 2: Search Query Keyword Analysis
import string
raw_input = "Buy mobile phone buy phone online"
clean_text = raw_input.lower().translate(str.maketrans("", "", string.punctuation))
tokens = clean_text.split()
tally = {}
for word in tokens:
    tally[word] = tally.get(word, 0) + 1
repeated_terms = {k: v for k, v in tally.items() if v > 1}
print(repeated_terms)
```

```python
# Problem 3: Sensor Data Validation
telemetry_log = [3, 4, 7, 8, 10, 12, 5]
valid_entries = []
for timestamp, reading in enumerate(telemetry_log):
    if reading % 2 == 0:
        valid_entries.append((timestamp, reading))
print("Valid Sensor Readings (Hour, Value):")
print(valid_entries)

```

```python


# Problem 4: Email Domain Usage Analysis
user_emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]
domain_map = {}
for entry in user_emails:
    provider = entry.split("@")[-1]
    domain_map[provider] = domain_map.get(provider, 0) + 1
total_count = len(user_emails)
for domain, freq in domain_map.items():
    ratio = (freq / total_count) * 100
    print(f"{domain}: {ratio:.0f}%")

```

```python
# Problem 5: Sales Spike Detection
revenue_data = [1200, 1500, 900, 2200, 1400, 3000]
avg_revenue = sum(revenue_data) / len(revenue_data)
surge_limit = avg_revenue * 1.3
for day_num, amount in enumerate(revenue_data, 1):
    if amount > surge_limit:
        print(f"Day {day_num}: {amount}")

```

```python
# Problem 6: Duplicate User ID Detection
id_registry = ["user1", "user2", "user1", "user3", "user1", "user3"]
occurrence_map = {}
for uid in id_registry:
    occurrence_map[uid] = occurrence_map.get(uid, 0) + 1
for uid, count in occurrence_map.items():
    if count > 1:
        print(f"{uid} → {count} times")
```
