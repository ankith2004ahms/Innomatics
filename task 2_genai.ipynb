```python
## Problem 1: Unique Words in a Sentence

line_of_words = input("Enter the sentence: ")

words = line_of_words.split()
unique_words = set(words)

print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
```

```python
## Highest Salary from Employee Data
corporate_salary_matrix = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}
salary_peak_holder = max(corporate_salary_matrix, key=corporate_salary_matrix.get)

print("Highest Salary:", salary_peak_holder, "-", corporate_salary_matrix[salary_peak_holder])
```

```python
## Find Maximum and Minimum Values
numeric_sequence_container = [45, 22, 89, 10, 66]
upper_bound_tracker = numeric_sequence_container[0]
lower_bound_tracker = numeric_sequence_container[0]
for numeric_unit in numeric_sequence_container:
    if numeric_unit > upper_bound_tracker:
        upper_bound_tracker = numeric_unit
    if numeric_unit < lower_bound_tracker:
        lower_bound_tracker = numeric_unit
print("List:", numeric_sequence_container)
print("Max:", upper_bound_tracker)
print("Min:", lower_bound_tracker)
```

```python
## Count Products Above a Price Threshold
retail_price_catalog = [450, 1200, 899, 1500, 300]
high_value_item_counter = 0
for price_token in retail_price_catalog:
    if price_token > 1000:
        high_value_item_counter += 1

print("Products above 1000:", high_value_item_counter)
```

```python
##Calculate Attendance Percentage
presence_record_series = ["P", "P", "A", "P", "P"]
present_day_counter = presence_record_series.count("P")
total_session_days = len(presence_record_series)
attendance_percentage_value = (present_day_counter / total_session_days) * 100
print("Attendance Percentage:", attendance_percentage_value)
```

```python
##Remove Duplicate Phone Numbers
mobile_contact_archive = [9876543210, 9123456789, 9876543210]
unique_contact_registry = set(mobile_contact_archive)
print("Unique phone numbers:", unique_contact_registry)
```

```python
##Count Character Frequency
character_input_payload = "python"
character_frequency_index = {}
for letter_unit in character_input_payload:
    if letter_unit in character_frequency_index:
        character_frequency_index[letter_unit] += 1
    else:
        character_frequency_index[letter_unit] = 1
print(character_frequency_index)
```
