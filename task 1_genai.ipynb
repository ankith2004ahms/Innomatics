```python
# Task 1: User Login Check

username = "admin"
password = "1234"

username = input("Enter username: ")
u_password = input("Enter password: ")

if username == username and u_password == password:
    print("Login Successful")
else:
    print("Invalid Credentials")

```

```python
# Task 2: Pass / Fail Analyzer

marks = [45, 78, 90, 33, 60]

p_count = 0
fail_count = 0

for mark in marks:
    if mark >= 50:
        p_count += 1
    else:
        fail_count += 1

print("Total Students:", len(marks))
print("Pass Students:", p_count)
print("Fail Students:", fail_count)

```

```python
# Task 3: Simple Data Cleaner

names = [" Alice ", "bob", " CHARLIE "]

cleaned_names = []

for name in names:
    cleaned_names.append(name.strip().lower())

print("Cleaned Names List:", cleaned_names)


```

```python
# Task 4: Message Length Analyzer

messages = ["Hi", "Welcome to the platform", "OK"]

for msg in messages:
    print("Message:", msg)
    print("Length:", len(msg))

    if len(msg) > 10:
        print("⚠️ Flag: Message longer than 10 characters")

    print("--------------------")
```

```python
# Task 5: Error Message Detector

logs = ["INFO", "ERROR", "WARNING", "ERROR"]

error_count = 0

for log in logs:
    if log == "ERROR":
        error_count += 1

print("Total ERROR entries:", error_count)
```
